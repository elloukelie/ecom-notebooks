```python
import matplotlib.pyplot  as plt
import matplotlib.image  as mpimg
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
image = np.array([
    [
        [255, 0, 0], [0, 255, 0]
    ],
    [
        [0, 0, 255], [255, 255, 255]
    ]
    ], dtype=np.uint8)
plt.imshow(image)
plt.show()
```

```python
image_as_array = mpimg.imread('csvs/palm_trees.jpg')
image_as_array
```

```python
image_as_array.shape
```

```python
(h, w, c) = image_as_array.shape
reshaped_image_arr = image_as_array.reshape(h*w, c)
reshaped_image_arr
```

```python
model = KMeans(n_clusters=6)
labels = model.fit_predict(reshaped_image_arr)
labels
```

```python
rgb_codes = model.cluster_centers_
rgb_codes
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
rgb_codes
```

```python
rgb_codes[labels]
```

```python
quantized_image = np.reshape(rgb_codes[labels], (h,w, c))
quantized_image
```

```python
# og image
plt.imshow(image_as_array)
```

```python
# quantized image
plt.imshow(quantized_image)
```

```python
corgi_as_array = mpimg.imread('csvs/corgi.jpg')
```

```python
# Represent the image as a 3D array and find the image height and width values
corgi_as_array.shape
```

```python
# Perform image quantization using K-Mean Clustering model
# Check the image quantization result for → K = 4, K = 10, K = 25 and K = 50
```

```python
(h, w, c) = corgi_as_array.shape
reshaped_corgi_as_array = corgi_as_array.reshape(h*w, c)
reshaped_corgi_as_array
```

```python
model = KMeans(n_clusters=4)
corgi_labels = model.fit_predict(reshaped_corgi_as_array)
rgb_codes = model.cluster_centers_.round(0).astype(int)
quantized_corgi = np.reshape(rgb_codes[corgi_labels], (h,w, c))
plt.imshow(quantized_corgi)
```

```python
model = KMeans(n_clusters=10)
corgi_labels = model.fit_predict(reshaped_corgi_as_array)
rgb_codes = model.cluster_centers_.round(0).astype(int)
quantized_corgi = np.reshape(rgb_codes[corgi_labels], (h,w, c))
plt.imshow(quantized_corgi)
```

```python
model = KMeans(n_clusters=25)
corgi_labels = model.fit_predict(reshaped_corgi_as_array)
rgb_codes = model.cluster_centers_.round(0).astype(int)
quantized_corgi = np.reshape(rgb_codes[corgi_labels], (h,w, c))
plt.imshow(quantized_corgi)
```

```python
model = KMeans(n_clusters=50)
corgi_labels = model.fit_predict(reshaped_corgi_as_array)
rgb_codes = model.cluster_centers_.round(0).astype(int)
quantized_corgi = np.reshape(rgb_codes[corgi_labels], (h,w, c))
plt.imshow(quantized_corgi)
```
